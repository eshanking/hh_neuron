```python
"""
Math 449, final paper part I

Review of neuron action potential

In neuron.py, I coded from scratch a HH simulation of neuron action potential. In this notebook,
I use my code to explore some dynamical behavior of the HH model. First, I demonstrate basic 
spiking behavior given an injection current. Then, I demonstrate sub-threshold behavior given a
very small injection current. Finally, I demonstrate repeated spiking and show how different 
injection currents restult in different spiking rates. Voltage timecourse and phase-plan plots
are shown.

code available: https://github.com/eshanking/hh_neuron
"""
from neuron import Neuron
import numpy as np
import matplotlib.pyplot as plt

# simple spiking
inj = np.zeros(1*10**5)
for i in range(1*10**3):
    inj[i+2000] = 20

n1 = Neuron(inj=inj)
n1.simulate(1*10**5)
f,a = n1.plot()
f.savefig('basic_spiking.pdf')
```

```python
# sub-threshold
inj = np.zeros(1*10**5)
for i in range(1*10**3):
    inj[i+2000] = 1
n1 = Neuron(inj=inj)
n1.simulate(1*10**5)
f,a = n1.plot()
f.savefig('sub_thresh.pdf')
```

```python
# repeated spiking
inj = np.zeros(1*10**5)
for i in range(8*10**4):
    inj[i+5000] = 20

n1 = Neuron(inj=inj)
n1.simulate(1*10**5)
f,a = n1.plot()
f.savefig('repeated_spiking.pdf')
```

```python
# slow repeated spiking
inj = np.zeros(1*10**5)
for i in range(8*10**4):
    inj[i+5000] = 8

n1 = Neuron(inj=inj)
n1.simulate(1*10**5)
f,a = n1.plot()
f.savefig('repeated_spiking_slow.pdf')
```

```python
# fast repeated spiking
inj = np.zeros(1*10**5)
for i in range(8*10**4):
    inj[i+5000] = 50

n1 = Neuron(inj=inj)
n1.simulate(1*10**5)
f,a = n1.plot()
f.savefig('repeated_spiking_fast.pdf')
```
